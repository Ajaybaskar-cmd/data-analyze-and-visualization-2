```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tip=sns.load_dataset('tips')
tip.head()
```

```python
sns.set_style('white')
sns.countplot(x='sex',data=tip,palette='rainbow',hue='sex')
```

```python
sns.set_style('dark')
sns.countplot(x='sex',data=tip)
```

```python
sns.set_style('whitegrid')
sns.countplot(x='sex',data=tip)
```

```python
sns.set_style('ticks')
sns.countplot(x='sex',data=tip)
```

```python
plt.figure(figsize=(4,4))
sns.countplot(x='sex',data=tip)
```

```python
sns.set_context('poster',font_scale=1)
sns.countplot(x='sex',data=tip)
```
