```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
%matplotlib inline
```

```python
df=sns.load_dataset('tips')
df.head()
```

```python
#bar plot
sns.barplot(x='sex',y='total_bill',data=df)
```

```python
#countplot
sns.countplot(x='sex',data=df)
```

```python
sns.set_style('whitegrid')
sns.boxplot(x='sex',y='total_bill',data=df)
```

```python
sns.boxplot(data=df,orient='h')
```

```python
sns.violinplot(x='day',y='total_bill',data=df,palette='rainbow')
```

```python
sns.violinplot(x='day',y='total_bill',data=df,palette='rainbow',hue='smoker')
```

```python
sns.stripplot(x='day',y='total_bill',data=df,palette='rainbow',hue='sex',jitter=False,dodge=True)
```

```python
sns.swarmplot(x='day',y='total_bill',data=df,hue='sex')
```

```python
sns.violinplot(x='day',y='total_bill',data=df,palette='rainbow')
sns.swarmplot(x='day',y='total_bill',data=df,color='black',size=3)
```

```python
sns.catplot(x='day',y='total_bill',data=df,kind='violin')
```
